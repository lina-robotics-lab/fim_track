```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML

import pickle as pkl
import networkx as nx
```

```python
from matplotlib.lines import Line2D

fig, axes = plt.subplots(2,dpi=150,sharex=True)
   
for gain in  [0,0.1,0.2]:
    expr= "CoordDescent-F_gain{}-v1".format(gain)
    filepath = "{}.pkl".format(expr)
    with open(filepath,'rb') as file:
        rep_record = pkl.load(file)

    # print(rep_record)
    #     title=r"Consensus Gain={}".format(gain)

    qs = np.array([record['src_locs']['source_1'] for record in rep_record])
    # qs.shape = (n_expr,n_step,q_dim)

 
#     plt.suptitle('Consensus Scheme 1. FIM Consensus Gain={}'.format(gain))

    ax = axes[0]
    est = np.array([list(rec['est_locs_log'].values()) for rec in rep_record])
    err = np.median(np.linalg.norm(est-qs[:,np.newaxis,:,:],axis=-1),axis=1)
    err_ave = np.mean(err,axis=0)
    err_sd = np.std(err,axis=0)
    t = np.arange(len(err_ave))
    ax.plot(t,err_ave,label='Gain={}'.format(gain))
#     ax.fill_between(t,err_ave-err_sd,err_ave+err_sd,alpha=0.3,label='Gain={}'.format(gain))
    ax.set_ylabel('Est Error')

    ax = axes[1]
    loc = np.array([list(rec['sensor_locs'].values()) for rec in rep_record])
    dist = np.median(np.linalg.norm(loc-qs[:,np.newaxis,:,:],axis=-1),axis=1)
    dist_ave = np.mean(dist,axis=0)
    dist_sd = np.std(dist,axis=0)
    t = np.arange(len(dist_ave))
    ax.plot(t,dist_ave,label='Gain={}'.format(gain))
#     ax.fill_between(t,dist_ave-dist_sd,dist_ave+dist_sd,alpha=0.3,label='Gain={}'.format(gain))
    ax.set_ylabel("Sensor-Src Dist")


plt.suptitle('Consensus Scheme 1')

plt.legend()

plt.savefig('Coordination v1.png')
plt.show()

```

```python
for gain in  [0.1,0.2]:
    fig, axes = plt.subplots(2,dpi=150,sharex=True)
    
    expr = 'CoordDescent-F_gain0-v2'
    filepath = "{}.pkl".format(expr)
    with open(filepath,'rb') as file:
        rep_record = pkl.load(file)

    qs = np.array([record['src_locs']['source_1'] for record in rep_record])
    ax = axes[0]
    est = np.array([list(rec['est_locs_log'].values()) for rec in rep_record])
    err = np.median(np.linalg.norm(est-qs[:,np.newaxis,:,:],axis=-1),axis=1)
    err_ave = np.mean(err,axis=0)
    err_sd = np.std(err,axis=0)
    t = np.arange(len(err_ave))
    ax.plot(t,err_ave,label='Local FIM',marker='.')
#         ax.fill_between(t,err_ave-err_sd,err_ave+err_sd,alpha=0.1)
    ax.set_ylabel('Est Error')

    ax = axes[1]
    loc = np.array([list(rec['sensor_locs'].values()) for rec in rep_record])
    dist = np.median(np.linalg.norm(loc-qs[:,np.newaxis,:,:],axis=-1),axis=1)
    dist_ave = np.mean(dist,axis=0)
    dist_sd = np.std(dist,axis=0)
    t = np.arange(len(dist_ave))
    ax.plot(t,dist_ave,label='Local FIM',marker='.')
    ax.fill_between(t,dist_ave-dist_sd,dist_ave+dist_sd,alpha=0.1)
    ax.set_ylabel("Sensor-Src Dist")
    
    
    for v in range(4):
        expr= "CoordDescent-F_gain{}-v{}".format(gain,v+1)
        filepath = "{}.pkl".format(expr)
        with open(filepath,'rb') as file:
            rep_record = pkl.load(file)

        # print(rep_record)
        #     title=r"Consensus Gain={}".format(gain)

        qs = np.array([record['src_locs']['source_1'] for record in rep_record])
        # qs.shape = (n_expr,n_step,q_dim)


        ax = axes[0]
        est = np.array([list(rec['est_locs_log'].values()) for rec in rep_record])
        err = np.median(np.linalg.norm(est-qs[:,np.newaxis,:,:],axis=-1),axis=1)
        err_ave = np.mean(err,axis=0)
        err_sd = np.std(err,axis=0)
        t = np.arange(len(err_ave))
        ax.plot(t,err_ave,label='v={}'.format(v+1))
        ax.fill_between(t,err_ave-err_sd,err_ave+err_sd,alpha=0.1)
        ax.set_ylabel('Est Error')

        ax = axes[1]
        loc = np.array([list(rec['sensor_locs'].values()) for rec in rep_record])
        dist = np.median(np.linalg.norm(loc-qs[:,np.newaxis,:,:],axis=-1),axis=1)
        dist_ave = np.mean(dist,axis=0)
        dist_sd = np.std(dist,axis=0)
        t = np.arange(len(dist_ave))
        ax.plot(t,dist_ave,label='v={}'.format(v+1))
        ax.fill_between(t,dist_ave-dist_sd,dist_ave+dist_sd,alpha=0.1)
        ax.set_ylabel("Sensor-Src Dist")
    
    
    plt.suptitle('FIM Consensus Gain={}'.format(gain))

    plt.legend()
    plt.show()
```
